```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
st_df = pd.read_csv('/Users/jc33/google_drive/thesis/analysis/files/source_tracker/st_out/sink_predictions.txt', sep='\t', index_col=0)
map_df = pd.read_csv('/Users/jc33/google_drive/thesis/analysis/files/initial_steps/mapping_files/periods1-2-3_all_data_barcodes_map141108.txt', sep='\t', index_col=0)
st_df = st_df.join(map_df, how='inner')
```

```python
flag_office_1 = st_df[(st_df['City'] == 'flagstaff') & (st_df['OfficeNumber'] == 1)]
flag_office_1['Days'] = flag_office_1['DaysSinceEpoch'] - 15881 
flag_office_2 = st_df[(st_df['City'] == 'flagstaff') & (st_df['OfficeNumber'] == 2)]
flag_office_2['Days'] = flag_office_2['DaysSinceEpoch'] - 15881 
flag_office_3 = st_df[(st_df['City'] == 'flagstaff') & (st_df['OfficeNumber'] == 3)]
flag_office_3['Days'] = flag_office_3['DaysSinceEpoch'] - 15881 

toronto_office_2 = st_df[(st_df['City'] == 'toronto') & (st_df['OfficeNumber'] == 2)]
toronto_office_2['Days'] = toronto_office_2['DaysSinceEpoch'] - 15881 
toronto_office_1 = st_df[(st_df['City'] == 'toronto') & (st_df['OfficeNumber'] == 1)]
toronto_office_1['Days'] = toronto_office_1['DaysSinceEpoch'] - 15881
toronto_office_3 = st_df[(st_df['City'] == 'toronto') & (st_df['OfficeNumber'] == 3)]
toronto_office_3['Days'] = toronto_office_3['DaysSinceEpoch'] - 15881

san_diego_office_1 = st_df[(st_df['City'] == 'san_diego') & (st_df['OfficeNumber'] == 1)]
san_diego_office_1['Days'] = san_diego_office_1['DaysSinceEpoch'] - 15881
san_diego_office_2 = st_df[(st_df['City'] == 'san_diego') & (st_df['OfficeNumber'] == 2)]
san_diego_office_2['Days'] = san_diego_office_2['DaysSinceEpoch'] - 15881
san_diego_office_3 = st_df[(st_df['City'] == 'san_diego') & (st_df['OfficeNumber'] == 3)]
san_diego_office_3['Days'] = san_diego_office_3['DaysSinceEpoch'] - 15881
```
