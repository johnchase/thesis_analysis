```python
# Create a list of all the ids to be compared
# pull out the distances and the time, given by days since epoch - days since epoch
def get_all_distance_pairs(df, dm, date_col, comparison_col, columns_to_include):
    #make sure to get the overlap of ids to prevent errors
    dm_ids = dm.ids
    df_ids = df.index
    master_ids = df_ids.intersection(dm_ids)
    
    
    dm = dm.filter(master_ids)
    df = df.ix[master_ids]
    #Create an empty data frame with the columns that need to be added
    distances_df = pd.DataFrame()
    distances_df[date_col] = np.nan
    distances_df['Distances'] = np.nan
    distances_df['IDs'] = np.nan
    for column in columns_to_include:
        distances_df[column] = np.nan
        
        
    for comparison_id in set(df[comparison_col]):
        sample_ids = df[df[comparison_col] == comparison_id].index
        sample_id_pairs_lookup = []
        for j in sample_ids:
            for k in sample_ids:
                if j == k or (k,j) in sample_id_pairs_lookup:
                    pass
                else:
                    id_pair = ''.join([j, k])
                    distances_df.loc[id_pair] = np.nan
                    distances_df['IDs'][id_pair] = id_pair
                    distances_df['Distances'][id_pair] = dm[j, k]
                    distances_df[date_col][id_pair] = abs(df.loc[j, date_col] - df.loc[k, date_col])
                    for column in columns_to_include:
                        distances_df[column][id_pair] = df[column][j]
    return distances_df
```

```python
def get_polyfit_with_R2(x, y, degree):
    #potentially confusing functionn name, use 
    results = {}
    coeffs = np.polyfit(x, y, degree)
    
    # Polynomial Coefficients
    results['polynomial'] = coeffs.tolist()

    p = np.poly1d(coeffs)
    # fit values, and mean
    yhat = p(x)
    ybar = np.sum(y)/len(y)
    ssreg = np.sum((yhat-ybar)**2)
    sstot = np.sum((y - ybar)**2)
    # r-squared
    results['determination'] = ssreg / sstot
    return results

# ("Days", "Distances", day_dist_df, order=2);
x = day_dist_df['Days']
y = day_dist_df['Distances']

results = get_polyfit_with_R2(x, y, 2)
print str(round(results['determination'], 2))
```

```python
# Create a list of all the ids to be compared
# pull out the distances and the time, given by days since epoch - days since epoch
def get_all_distance_pairs(df, dm, date_col, id_col):
    distances = []
    sample_id_pairs = []
    day_changes = []
    no_event_ids = set(df[id_col])
    for i in no_event_ids:
        sample_ids = df[df[id_col] == i].index
        sample_id_pairs_lookup = []
        for j in sample_ids:
            for k in sample_ids:
                if j == k or (k,j) in sample_id_pairs_lookup:
                    pass
                else:
                    id_pair = (j, k)
                    distances.append(dm[id_pair])
                    day_change = abs(df.loc[j, date_col] - df.loc[k, date_col])
                    day_changes.append(day_change)
                    sample_id_pairs.append(id_pair)
                    sample_id_pairs_lookup.append(id_pair)
    return distances, day_changes

```
