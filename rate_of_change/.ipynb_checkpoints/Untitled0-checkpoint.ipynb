```python
# we need to get the intersection of the ids from the mapping file and the distance matrix to avoid errors. The reason there
#are ids in the mapping file not in the dm is because the dm has been filtered to a depth of 800 ses per sample.
flag_df.set_index('#SampleID', inplace=True)
dm_ids = weighted_dm.ids
map_ids = flag_df.index
master_ids = map_ids.intersection(dm_ids)
```

```python
filtered_weighted_dm = weighted_dm.filter(master_ids)
filtered_flag_df = flag_df.ix[master_ids]
```

```python
# Create a list of all the ids to be compared
# pull out the distances and the time, given by days since epoch - days since epoch
def get_all_distance_pairs(df, dm, date_col, id_col):
    distances = []
    sample_id_pairs = []
    day_changes = []
    no_event_ids = set(df[id_col])
    for i in no_event_ids:
        sample_ids = df[df[id_col] == i].index
        sample_id_pairs_lookup = []
        for j in sample_ids:
            for k in sample_ids:
                if j == k or (k,j) in sample_id_pairs_lookup:
                    pass
                else:
                    id_pair = (j, k)
                    distances.append(dm[id_pair])
                    day_change = abs(df.loc[j, date_col] - df.loc[k, date_col])
                    day_changes.append(day_change)
                    sample_id_pairs.append(id_pair)
                    sample_id_pairs_lookup.append(id_pair)
    return distances, day_changes

```

```python
distances, days = get_all_distance_pairs(filtered_flag_df, filtered_weighted_dm, 'DaysSinceEpoch', 'NoEvent')
```

```python
day_change_distance_df = pd.DataFrame(zip(distances, days), columns=['Distances', 'Days'])
```
